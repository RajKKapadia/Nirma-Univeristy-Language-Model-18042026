## Why character-level text?

Character-level modeling is useful for teaching because:
- the vocabulary is small
- preprocessing is simple
- students can clearly see next-character prediction
- generated text makes model differences visible

Example:

Input: `HELLO`
Target: `ELLO!`

## Vanilla RNN idea

At each time step:
- input: current character embedding
- hidden state: memory from previous time step
- output: prediction for next character

The hidden state acts like short memory.
The same weights are reused at every time step.

## Limitation of vanilla RNN

Main issue:
- it struggles to remember information over long sequences

Why?
- during backpropagation, gradients can become very small
- this is called the vanishing gradient problem

## GRU intuition

GRU adds gates to control memory flow.

Two important ideas:
- update gate: how much past information to keep
- reset gate: how much past information to forget

## LSTM intuition

LSTM introduces a more explicit memory cell.

Common gates:
- forget gate
- input gate
- output gate

In the code cell below, we build a very simple comparison chart. It is not a mathematically precise benchmark. Its purpose is classroom communication: to give students a quick visual summary that RNNs are simplest, GRUs add moderate complexity, and LSTMs are usually the most expressive for long-term memory.

```python
import matplotlib.pyplot as plt

models = ["RNN", "GRU", "LSTM"]
complexity = [1, 2, 3]
long_memory = [1, 2, 3]

plt.figure(figsize=(8, 4))
plt.plot(models, complexity, marker="o", label="Complexity")
plt.plot(models, long_memory, marker="o", label="Long-term memory ability")
plt.title("High-Level Comparison")
plt.ylabel("Relative score")
plt.legend()
plt.tight_layout()
plt.show()
```
