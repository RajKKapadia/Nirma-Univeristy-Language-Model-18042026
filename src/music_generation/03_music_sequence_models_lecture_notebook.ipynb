# Music Sequence Models — Lecture Notebook

This notebook is designed for classroom teaching.
It uses symbolic melodies to explain how sequence models can generate structured temporal patterns.

## Learning goals

By the end of this lecture, students should understand:
1. why music can be treated as sequence data
2. how note-duration tokens replace raw audio samples
3. how next-token prediction works for melody generation
4. why vanilla RNNs struggle with longer context
5. how GRU and LSTM improve sequence memory

## Why symbolic music instead of raw audio?

For teaching, symbolic music is better because:
- the vocabulary is small enough to inspect
- preprocessing is simple
- each token has a visible musical meaning
- the sequence-model idea stays clear

Example melody:

`C4_q D4_q E4_q C4_q`

## What is the model predicting?

At each time step the model reads one token and predicts the next token.

Example:
- input: `C4_q D4_q E4_q`
- target: `D4_q E4_q C4_q`

This is the same next-step prediction idea used in text generation.

## Vanilla RNN intuition

A vanilla RNN processes the melody left to right.
- current token enters the model
- hidden state carries previous context
- output distribution predicts the next note event

The hidden state acts like short-term musical memory.

## Why vanilla RNN can struggle

Music often depends on patterns over several steps:
- repeated motifs
- phrase endings
- call-and-response structure

Vanilla RNNs can forget earlier context because of vanishing gradients.

## GRU and LSTM intuition

GRU and LSTM introduce gates that help decide:
- what to keep from the past
- what to ignore
- when to expose memory to the output

That usually makes them better at longer melodic structure than a plain RNN.

```python
import matplotlib.pyplot as plt

models = ["RNN", "GRU", "LSTM"]
complexity = [1, 2, 3]
memory = [1, 2, 3]

plt.figure(figsize=(8, 4))
plt.plot(models, complexity, marker="o", label="Relative complexity")
plt.plot(models, memory, marker="o", label="Long-range memory ability")
plt.title("High-Level Sequence Model Comparison")
plt.ylabel("Relative score")
plt.legend()
plt.tight_layout()
plt.show()
```

## Suggested teaching flow

1. show a few raw melody lines

3. build a vocabulary and encode tokens as integers
4. create input-target sequence windows
5. train a vanilla RNN
6. compare it with a GRU and an LSTM
7. listen to generated samples and discuss structure
