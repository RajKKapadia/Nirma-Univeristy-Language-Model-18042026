## What is different from character-level generation?

In the character-level notebook the model predicts the next symbol at every time step.

In sentiment analysis:
- the full sentence is the input sequence
- the output is a single label such as `positive` or `negative`
- the model must compress the meaning of the sequence into a final representation for classification

## Why use sequence models for sentiment?

Sentiment often depends on word order.

Examples:
- `good` and `not good` should not receive the same prediction
- `surprisingly moving ending` carries a different tone from `surprisingly boring ending`

Sequence models process tokens in order and can learn these context effects better than a naive bag-of-words approach.

## Vanilla RNN idea

At each time step the RNN reads one token embedding and updates its hidden state.

For sentiment classification:
- early tokens influence the hidden state
- later tokens continue updating that state
- the final state is used to predict the sentiment label

## Limitation of vanilla RNN

A plain RNN can struggle to preserve important information across many time steps.

This matters for sentiment because:
- the key opinion word may appear late in the sentence
- negation may flip the meaning of earlier words
- gradients can vanish during training

## GRU and LSTM intuition

GRU and LSTM introduce gates that control what information is kept or forgotten.

High-level intuition:
- GRU is simpler and often easier to train than a vanilla RNN
- LSTM uses a more explicit memory pathway and can preserve useful context over longer spans
- both are common upgrades when sequence order matters

```python
import matplotlib.pyplot as plt

models = ["RNN", "GRU", "LSTM"]
complexity = [1, 2, 3]
memory_support = [1, 2, 3]

plt.figure(figsize=(8, 4))
plt.plot(models, complexity, marker="o", label="Relative complexity")
plt.plot(models, memory_support, marker="o", label="Long-range memory support")
plt.title("High-Level Comparison for Sentiment Classification")
plt.ylabel("Relative score")
plt.legend()
plt.tight_layout()
plt.show()
```
